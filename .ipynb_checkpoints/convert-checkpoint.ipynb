```python
import pandas as pd 
from sqlalchemy import text,create_engine
```

```python
engin=create_engine('sqlite:///us_cunses_data.db')
```

```python
df=pd.read_csv('US_Census_Data.csv')
```

```python
df.to_sql('Census_Data',engin,index=False,if_exists='replace')
```

```python
Census_Data=pd.read_sql(text('select * from Census_Data'),con=engin.connect())
```
